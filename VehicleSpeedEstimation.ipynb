# Step 1: Environment & Library Setup


```python
!pip install ultralytics==8.2.103 opencv-python -q
from ultralytics import YOLO
from ultralytics.solutions import speed_estimation
import cv2
```

# Step 2: Load Model & Video Setup


```python
model = YOLO("yolov8n.pt")
cap = cv2.VideoCapture("/content/drive/MyDrive/Bilgisayar mühendisliği/Yapay Zeka/YOLOv8: Başlangıç/2. Speed Estimation/Video/cars.mp4")
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH, cv2.CAP_PROP_FRAME_HEIGHT, cv2.CAP_PROP_FPS))
video_writer = cv2.VideoWriter("speed_estimation.avi", cv2.VideoWriter_fourcc(*'mp4v'), fps, (w, h))
```

# Step 3: Speed Estimation Setup

```python
line_pts = [(0, 200), (1280, 200)]
speed_obj = speed_estimation.SpeedEstimator(names=model.names, reg_pts=line_pts, view_img=True)
```

# Step 4: Process Video and Estimate Speed

```python
while cap.isOpened():
    success, im0 = cap.read()
    if not success:
        print("Video frame is empty or processing complete.")
        break

    tracks = model.track(im0, persist=True, show=False)
    im0 = speed_obj.estimate_speed(im0, tracks)
    video_writer.write(im0)

cap.release()
video_writer.release()
cv2.destroyAllWindows()
```
